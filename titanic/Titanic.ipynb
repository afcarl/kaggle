```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
def transform_data(csv):
    df = pd.read_csv(csv)
    df = df.fillna({'Age': -1, 'Cabin': 'Unknown', 'Embarked': 'Unknown'})
    df['CabinLetter'] = df['Cabin'].str.slice(0,1)
    df['ParchGT0'] = df['Parch'] > 0
    df['SibSpGT0'] = df['SibSp'] > 0
    df['Name'] = df['Name'].str.lower()
    df['mr'] = df['Name'].str.contains('mr\.')
    df['mrs'] = df['Name'].str.contains('mrs\.')
    df['miss'] = df['Name'].str.contains('miss')
    df['master'] = df['Name'].str.contains('master')
    df['jr'] = df['Name'].str.contains('jr')
    df = pd.get_dummies(df, columns=['Pclass', 'Cabin', 'Embarked', 'CabinLetter', 'Sex'])
    
    vectorizer = CountVectorizer(min_df=.01)
    
    df = pd.concat([
        df,
        pd.DataFrame(vectorizer.fit_transform(df['Ticket']).toarray())
    ], axis=1)
    
    return df.loc[:,(df.dtypes =='int64') | (df.dtypes =='bool') | (df.dtypes =='float64')]


def transform_data(csv):
    df = pd.read_csv(csv)
    df = df.fillna({'Age': -1, 'Cabin': 'Unknown', 'Embarked': 'Unknown'})
    cols = ['PassengerId', 'Age', 'Sex', 'Pclass', 'Cabin', 'Name']
    if 'train' in csv:
        cols = ['Survived'] + cols
    df = df[cols]
    
    #df['CabinLetter'] = df['Cabin'].str.slice(0,1)
    df['master'] = df['Name'].str.contains('master')
    df['mr'] = df['Name'].str.contains('mr\.')
    df['mrs'] = df['Name'].str.contains('mrs\.')
    df['miss'] = df['Name'].str.contains('miss')
    
    #df['ParchGT0'] = df['Parch'] > 0
    #df['SibSpGT0'] = df['SibSp'] > 0
    
    df = pd.get_dummies(df, columns=['Sex', 'Pclass'])
    
    return df.loc[:,(df.dtypes =='int64') | (df.dtypes =='bool') | (df.dtypes =='float64')]

train = transform_data('train.csv')
test = transform_data('test.csv')

#train

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

Y = train[['Survived']].values
X = train.values[:,2:]
clf = RandomForestClassifier(n_estimators=500)
clf.fit(X,Y)
print('train:', accuracy_score(Y, clf.predict(X)))

#X_test = test.values[:,1:]
#Y_test = test[['Survived']].values
#print('train:', accuracy_score(Y_test, clf.predict(X_test)))

cross_validate(clf, X, Y.reshape(-1))

X_test = test.values[:,1:]
#Y_test = test[['Survived']].values
#print('train:', accuracy_score(Y_test, clf.predict(X_test)))
test['Survived'] = clf.predict(X_test)
test[['PassengerId', 'Survived']]
```
